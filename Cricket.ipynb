```python
import pandas as pd
```

```python
#import CSV File

df = pd.read_csv('cricket.csv')
df
```

```python
#renaming the columns

df = df.rename(columns = {'NO' : 'Not-Out',
                          'BF' : 'Balls-Faced',
                          'Mat': 'Matches',
                          'HS' : 'Highest-Inns-Rate',
                          'SR' : 'Strike-Rate',
                          'Ave': 'Average'}
              )

df
            
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Adding Country Column

df['Country'] = df['Player'].str.split(pat = '(').str[1]
```

```python
df
```

```python
df['Country'] = df['Country'].str.split(pat = ')').str[0]
```

```python
df['Country'] 
```

```python
# Standardise the Player Column i.e Removing the Country Code

df['Player'] = df['Player'].str.split(pat = '(').str[0]
df
```

# Spliting the Span Column into Start and End Year

```python
df['Start-Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df['End-Year'] = df['Span'].str.split(pat = '-').str[1]
df
```

```python
# Droping the Column Span
```

```python
df = df.drop(['Span'], axis = 1)
```

```python
df
```

# Standardising Columns

```python
df['Matches'].str.split(pat = '*')
```

```python
df['Matches'] = df['Matches'].str.split(pat = '*').str[0]
```

```python
df
```

```python
df['Highest-Inns-Rate'].str.split(pat = '*')
```

```python
df['Highest-Inns-Rate'] = df['Highest-Inns-Rate'].str.split(pat = '*').str[0]
df
```

```python
df['Balls-Faced'] = df['Highest-Inns-Rate'].str.split(pat = '+').str[0]
df
```

# Questions 

```python
# Ques 1 - Find the Average Career Length
```

```python
df['Career-Length'] = df['End-Year'] - df['Start-Year']
```

```python
df['Career-Length'].mean()
```

```python
df
```

```python
# Ques 2 Find the average Strike-Rate of players who played over 10 years
```

```python
df[df['Career-Length'] > 10]['Strike-Rate'].mean()
```

```python
# Ques 3 Find the nomber of players who played before 1960
```

```python
df[df['Start-Year'] > 1960]['Player'].count()
```

```python
# Ques 4 Find the Highest Inns Score by country
```

```python
df.groupby('Country')['Highest-Inns-Rate'].max().to_frame('HIGHEST_INNS_SCORE').reset_index().sort_values('HIGHEST_INNS_SCORE', ascending = False)
```

```python
# Ques 5 Find the average of 100s, 50 s, 0s by Country
```

```python
df.groupby('Country')[['100', '50', '0']].mean()
```
